# 1. 편미분 함수를 등고선으로 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
#함수와 편미분함수 정의
def f(w0, w1):
    return w0**2+2*w0*w1 +3
def df_dw0(w0,w1):
    return 2*w0+2*w1
def df_dw1(w0,w1):
    return 2*w0+0*w1

w_range=2
dw=0.25
w0=np.arange(-w_range, w_range+dw,dw)
w1=np.arange(-w_range, w_range+dw,dw)
wn=w0.shape[0]
ww0,ww1=np.meshgrid(w0,w1)

ff=np.zeros((len(w0), len(w1)))
dff_dw0=np.zeros((len(w0),len(w1)))
dff_dw1=np.zeros((len(w0),len(w1)))

for i0 in range(wn):
    for i1 in range(wn):
        ff[i1,i0]=f(w0[i0], w1[i1])
        dff_dw0[i1,i0]=df_dw0(w0[i0],w1[i1])
        dff_dw1[i1,i0]=df_dw1(w0[i0],w1[i1])
        
plt.figure(figsize=(9,4))
cont=plt.contour(ww0,ww1,ff,10,color='k')
cont.clabel(fmt='%2.0f', fontsize=8)
plt.show()
```

# 2. 편미분함수를 경사벡터로 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
#함수와 편미분함수 정의
def f(w0, w1):
    return w0**2+2*w0*w1 +3
def df_dw0(w0,w1):
    return 2*w0+2*w1
def df_dw1(w0,w1):
    return 2*w0+0*w1

w_range=2
dw=0.25
w0=np.arange(-w_range, w_range+dw,dw)
w1=np.arange(-w_range, w_range+dw,dw)
wn=w0.shape[0]
ww0,ww1=np.meshgrid(w0,w1)

ff=np.zeros((len(w0), len(w1)))
dff_dw0=np.zeros((len(w0),len(w1)))
dff_dw1=np.zeros((len(w0),len(w1)))

for i0 in range(wn):
    for i1 in range(wn):
        ff[i1,i0]=f(w0[i0], w1[i1])
        dff_dw0[i1,i0]=df_dw0(w0[i0],w1[i1])
        dff_dw1[i1,i0]=df_dw1(w0[i0],w1[i1])
        
plt.figure(figsize=(9,4))
plt.quiver(ww0,ww1,dff_dw0, dff_dw1)
plt.show()
```

# 3. 편미분을 입체로 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#함수와 편미분함수 정의
def f(w0, w1):
    return w0**2+2*w0*w1 +3

w_range=2
dw=0.25
w0=np.arange(-w_range, w_range+dw,dw)
w1=np.arange(-w_range, w_range+dw,dw)

ww0,ww1=np.meshgrid(w0,w1)
z= f(ww0,ww1)


fig=plt.figure(figsize=(9,6))
ax=fig.gca(projection='3d')

surf=ax.plot_surface(ww0,ww1,z, cmap='coolwarm')
wire = ax.plot_wireframe(ww0,ww1,z,color='r',linewidth=0.1)
fig.colorbar(surf,shrink=0.5, aspect=5)

plt.show()

```
