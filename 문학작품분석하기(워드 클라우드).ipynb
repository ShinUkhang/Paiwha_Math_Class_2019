# 1. 홍길동전 단어정리

```python
#분석할 파일을 불러옵니다.
f = open("홍길동전.txt", "r")

from collections import Counter
wordDict = Counter()
sentences = f.readlines()

for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        wordDict[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
wordlist=[]
freqlist=[]
for word, freq in wordDict.most_common(40):
    wordlist.append(word)
    freqlist.append(freq)
    print(word,freq)
```

# 2. 홍길동전 워드클라우드

```python
from konlpy.tag import Hannanum
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = open("홍길동전.txt", "r").read() #encoding='utf8'는 오류가 나는 경우 사용

engin=Hannanum()
nouns=engin.nouns(text)
nouns=[n for n in nouns if len(n) >=2]

count=Counter(nouns)
tags=count.most_common(40)

wordcloud=WordCloud(font_path='c:/Windows/Fonts/malgun.ttf', background_color='white', width=1200, height=800).generate_from_frequencies(dict(tags))

fig=plt.figure()
plt.axis('off')
plt.imshow(wordcloud)
```

# 3. 홍길동전 명사형 정리

```python
from konlpy.tag import Hannanum
from collections import Counter

#분석할 파일을 불러옵니다.
f = open("홍길동전.txt", "r").read() 

engin=Hannanum()
nouns=engin.nouns(f)
nouns=[n for n in nouns if len(n) >=2]

count=Counter(nouns)
tags=count.most_common(40)

wordlist=[]
freqlist=[]
for word, freq in tags:
    wordlist.append(word)
    freqlist.append(freq)
    print(word,freq)
```
